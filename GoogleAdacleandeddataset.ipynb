```python
import pandas as pd
```

```python
df= pd.read_csv("/content/GoogleAds_DataAnalytics_Sales_Uncleaned.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Clicks'].fillna(df['Clicks'].median(), inplace=True)
df.isnull().sum()
```

```python
df[df['Impressions'].isnull()]
```

```python
df[['Impressions']]
```

```python
df['Impressions'] = df['Impressions'].fillna(method='bfill')
df.isnull().sum()
```

```python
print(df.duplicated())
```

```python
df['Leads'].fillna(0, inplace=True)
df['Conversions'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Ad_Date'] = pd.to_datetime(df['Ad_Date'], errors='coerce')
```

```python
df['Device'] = df['Device'].str.lower().str.strip()
```

```python
df['Location'] = df['Location'].str.lower().str.strip()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Cost'] = df['Cost'].replace({r'\$': ''}, regex=True).astype(float)
df['Cost'] = df['Cost'].fillna(df['Cost'].median())
```

```python
df['Sale_Amount'].fillna(0, inplace=True)
```

```python
df['Conversion Rate'] = df['Conversions'] / df['Clicks']
```

```python
df['Ad_Date'].fillna(method='bfill', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Ad_Date'], inplace=True)
```
